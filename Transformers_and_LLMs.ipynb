Transformer Architectures

*   Encoder Only
  * Use Cases: Sentiment Analysis, Classification,NER
  * Exapmle Models: BERT,RoBERTa...
*   Encoder-Decoder
  * Use Cases: Machine translation, Summarization, OCR
  * Example Models: BART, T5
*   Decoder Only
  * Use Cases: Generate text, images
  * Example Models: GPT-n, ChatGPT, PaLM, LLaMa



**How the LLMs are trained**



*   Step -1 (create **Foundation** Model)
  *   Create a Transformer model
  *   Gather whatever data you can find on internet
  *   Unsupervised training of n-gram data from above
*   Step-2 (Create **FLAN** Model)
  *   Supervised finetuning with control data
  *   Define PPO policy( Proximal Policy Optimization)
  *   Collect comparision choices and develop reward model
  *   Optimize Policy by training reward model using reinforcement learning

**What is Task Specific Fine-tuning**


*   Instruction Fine Tuning (IFT)
  * Expensive
  * All Model parameters modified
  * Catastrophic forgetting
*   Perameter Efficient Fine Tuning (PEFT)
  * Only few (new) parameres gets modified
  * Cost efficient
  * Original weights frozen




**How to Evaluate Models**
* ROUGE (Recall-Oriented Understudy for Gisting Evaluation)
  * how many the words (and/or n-grams) in the human reference summaries appeared in the machine generated summaries.

* BLEU (Bilingual Evaluation Understudy)
  * how many the words (and/or n-grams) in the machine generated summaries appeared in the human reference summaries.
